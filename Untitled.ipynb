```python
import pandas as pd
import numpy as np
```

```python
data= pd.read_csv('Hospital_Dataset.csv')
data.head(1)
```

```python
# rename columns properly
data = data.rename(columns= {'case_id' : 'case_id', 'Hospital_code': 'hospital_code', 'Hospital_region_code': 'hospital_region', 
                             'Available Extra Rooms in Hospital': 'rooms', 'Department': 'department', 'Ward_Type':'ward', 
                           'Ward_Facility_Code': 'Ward_Facility_Code', 'Bed Grade' : 'bed_grade', 'patientid': 'patient_id',
                           'City_Code_Patient': 'city_code_patient', 'Type of Admission': 'admission_type', 'Severity of Illness': 'illness_severity',
                            'Visitors with Patient': 'visitors_patient', 'age': 'age', 'Admission_Deposit': 'admission_deposit', 'Stay':'hospital_stay' })
```

```python
# the regions. data.head()
regions = data['hospital_region'].unique()
data['hospital_region'] = data['hospital_region'].replace({'X':'eastern', 'Y': 'western', 'Z':'southern'})
```

```python
# Hospital type
hospital_type = data['Hospital_type_code'].unique()
hospital_type_nbr = data['Hospital_type_code'].nunique()
data['Hospital_type_code'] = data['Hospital_type_code'].replace({'a':'health post', 'b': 'health center', 'c': 'medical health center',
                                                                 'd': 'district hospital', 'e': 'provoncial_hospital', 'f': 'teaching hospital',
                                                                 'g': 'refferal hospital', })
```

```python
# name wards using meaningful names
wards = data['ward'].unique()
data['ward'] = data['ward'].replace({'Q': 'medical', 'R': 'paediatric', 'S': 'ICU', 'T': 'OPD', 'P': 'surgical'})
data.head()
```

```python
# remove unnecessary columns
data = data.drop({'hospital_code', 'case_id', 'City_Code_Hospital', 'Ward_Facility_Code' })

data = data.drop({'hospital_code', 'case_id', 'City_Code_Hospital', 'Ward_Facility_Code'})
```

```python
# remove unnecessary columns
###### data = data.drop({'hospital_code', 'case_id', 'City_Code_Hospital', 'Ward_Facility_Code'}) ######
data = data.drop(columns=[col for col in ['hospital_code', 'case_id', 'City_Code_Hospital', 'Ward_Facility_Code'] if col in data.columns])
data.head()
```

```python
data.info()
```

```python
# there are missing values in in Two columns,
cat_col = []
num_col = []
for col in data.columns:
    if data[col].dtype == 'int64' or data[col].dtype == 'float64':
        num_col.append(col)
    elif data[col].dtype == 'object' or data[col].dtype.name == 'category':
        cat_col.append(col)

# lets impute missing values in categorical them using the mode
for col in num_col :
    mode_value = data[col].mode()[0]
    data[col].fillna(mode_value, inplace = True)

data.info()
```

```python
#### city code patient is behaving like a numerical while it is categorical, so we need to change its type to object.
data['city_code_patient'].dtype
data['city_code_patient'] = data['city_code_patient'].astype('object')
```

```python
data.to_csv()
data.to_csv('cleaned_data.csv', index=False)
```
